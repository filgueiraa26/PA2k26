```python
#1. Par ou ímpar
#Peça um número ao usuário e informe se ele é par ou ímpar.

numero = int(input("digite um numero: "))
num = numero % 2
if num == 0:
  print(f"o numero {numero} é par")
else:
  print(f" o numero {numero} é impar")
```

```python
#2. Maior de dois números
#Solicite dois números e mostre qual é o maior (ou se são iguais).
# entrada
a = int(input("digite um número: "))
b = int(input("digite outro número: "))
#processamento e saida
if a > b:
  print(f"{a} é maior que {b}")
elif a == b:
  print(f"{a} é igual a {b}")
else:
  print(f"{b} é maior que {a}")
```

```python
#3. Aprovação de aluno
#Peça a nota de um aluno e informe:
#Aprovado (nota ≥ 7)
#Recuperação (5 ≤ nota < 7)
#Reprovado (nota < 5)
nome = str(input("qual seu nome? "))
n1 = float(input(" qual sua nota do Primeiro bimestre? "))
if n1 > 10:
  print(" digite uma nota valida  (de 0 a 10)")
n2 = float(input(" qual sua nota do Segundo bimestre? "))
if n2 > 10:
  print(" digite uma nota valida  (de 0 a 10)")
n3 = float(input(" qual sua nota do Terceiro bimestre? "))
if n3 > 10:
  print(" digite uma nota valida  (de 0 a 10)")
n4 = float(input(" qual sua nota do Quarto bimestre? "))
if n4 > 10:
  print(" digite uma nota valida  (de 0 a 10)")
nf = n1 + n2 + n3 + n4
nb = nf / 4
if n1 > 10 or n2 > 10 or n3 > 10 or n4 > 10:
  print("digite notas validas (0 a 10)")
elif nb >= 7:
  print(f"o aluno {nome}  tirou {nb}  e foi aprovado")
elif nb >= 5:
    print(f"o aluno {nome} tirou {nb} e ficou de recuperação")
else:
    print(f"o aluno {nome} tirou {nb} foi reprovado")
```

```python
#4. Classificação de idade
#Receba a idade e classifique:
#Criança (0–12)
#Adolescente (13–17)
#Adulto (18–59)
#Idoso (60+)
idade = int(input("digite sua idade: "))
if idade <= 12:
  print("você é uma criança")
elif idade >= 13 and idade <= 17:
  print("você é um adolescente")
elif idade >= 18 and idade <= 59:
  print("você é um adulto")
else:
  print("você é um idoso")
```

```python
#6. Positivo, negativo ou zero
#Peça um número e diga se ele é positivo, negativo ou zero.
nu = float(input("Digite um número: "))
if nu > 0:
  print (f"o número {nu} é positivo.")
elif nu == 0:
    print(f"o numero {nu} é igual a zero")
else:
    print(f"o numero {nu} é negativo")
```

```python
#9.Ano bissexto
#Solicite um ano e informe se ele é bissexto ou não.
#Se for divisível por 4, exceto para anos múltiplos de 100 que não são múltiplos de 400.

ano = int(input("digite um ano: "))
if ano % 4 == 0:
  print(f"o ano {ano} é bissexto.")

else:
  print(f"o ano {ano} não é bissexto. ")
```

```python
#10.
#Desconto em compras
#Peça o valor de uma compra:
#≥ 100: 10% de desconto
#≥ 50 e < 100: 5% de desconto
#< 50: sem desconto
valor = float(input("qual o valor de sua compra? "))
if valor >= 100:
  desc = valor * 0.1
  prec = valor - desc
  print(f"sua compra ganhou 10% de desconto e ficou {prec}")
elif valor >= 50 and valor < 100:
  desc = valor * 0.05
  prec = valor - desc
  print(f"sua compra ganhou 5% de desconto e ficou {prec}")
else:
  print(f"sua compra não ganhou desconto, e ficou {valor}")
```
