```python
# Estrutura de  decisões (if, elf e else)

# Escrever um programa que comparar dois números

# Entradas
a = int (input("Digite um número inteiro: "))
b = int (input("Digite um número inteiro: "))

# Processamento / saída

if a == b:
  print (f"{a} é igual a {b}")
else:
  print (f"{a} é diferente de {b}")
```

```python
# Escrever um programa que verifica se o número digitado é par ou ímpar

# Entrada

numero = int (input("Digite um número: "))

# Processamento / Saída

if numero % 2 == 0:

  print (f" o numero {numero} é par. ")

else:
  print (f"O numero {numero} é ímpar. ")
```

```python
# Escrever um programa que p usuário digita o nome e a idade e verifica se o usuário é criança, adolescente ou adulto

# Entrada

nome = str  (input("digite seu nome: ."))
idade = int(input("digite sua idade:"))

# Processamento / saída

if idade < 12 :
  print (f"{nome} você é uma criança")

elif idade < 18:
  print (f"{nome} você é um Adolescente")

else:
  print (f"{nome} você é um Adulto")
```
